```python
# 1. Import required libraries
import seaborn as sns
import matplotlib.pyplot as plt
import pandas as pd
import numpy as np

# 2. Create synthetic data using NumPy
x = np.arange(1, 11)  # Numbers 1 to 10
y = np.random.randint(10, 100, size=10)  # Random values between 10 and 100

# 3. Create a DataFrame using Pandas
df = pd.DataFrame({
    'Day': x,
    'Sales': y
})

# 4. Plot using Seaborn
plt.figure(figsize=(8, 5))
sns.lineplot(data=df, x='Day', y='Sales', marker='o')

# 5. Customize the plot
plt.title('Daily Sales Over Time')
plt.xlabel('Day')
plt.ylabel('Sales')
plt.grid(True)

# 6. Show the plot
plt.show()
```

```python
# 1. Import libraries
import seaborn as sns
import matplotlib.pyplot as plt
import pandas as pd
import numpy as np

# 2. Create synthetic data
np.random.seed(42)
days = np.tile(np.arange(1, 11), 3)  # Days 110 repeated for 3 departments
sales = np.random.randint(50, 150, size=30)  # Random sales numbers
department = np.repeat(['Sales', 'IT', 'HR'], 10)  # 3 departments
region = np.random.choice(['North', 'South'], size=30)  # Random regions for style

# 3. Create DataFrame
df = pd.DataFrame({
    'Day': days,
    'Sales': sales,
    'Department': department,
    'Region': region
})

# 4. Create line plot
plt.figure(figsize=(10, 6))
sns.lineplot(data=df, x='Day', y='Sales', hue='Department', style='Region', markers=True)

# 5. Customize plot
plt.title('Department-wise Sales Over Days by Region')
plt.xlabel('Day')
plt.ylabel('Sales')
plt.grid(True)
plt.tight_layout()

# 6. Show plot
plt.show()
```

```python
# 1. Import libraries
import numpy as np
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt

# 2. Generate data using NumPy (normal distribution)
np.random.seed(42)
data = np.random.normal(loc=60, scale=15, size=1000)

# 3. Create a DataFrame using Pandas
df = pd.DataFrame({'Test_Scores': data})

# 4. Plot using Seaborn and Matplotlib
plt.figure(figsize=(10, 6))
sns.histplot(data=df, x='Test_Scores', kde=True, color='teal', bins=30)

# 5. Customize plot
plt.title('Distribution of Test Scores')
plt.xlabel('Test Scores')
plt.ylabel('Frequency')
plt.grid(True)
plt.tight_layout()

# 6. Show plot
plt.show()
```

```python
# 1. Import libraries
import numpy as np
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt

# 2. Generate data
np.random.seed(0)
x = np.random.randint(10, 100, size=50)
y = x + np.random.normal(0, 10, size=50)

# 3. Create DataFrame
df = pd.DataFrame({'X_values': x, 'Y_values': y})

# 5. Create scatter + line plot
plt.figure(figsize=(8, 5))
sns.scatterplot(data=df, x='X_values', y='Y_values', color='blue', label='Data Points')
sns.lineplot(data=df, x='X_values', y='Y_values', color='orange', label='Line Plot')

# 6. Customize the plot
plt.title('Scatter Plot with Line')
plt.xlabel('X Values')
plt.ylabel('Y Values')
plt.grid(True)
plt.legend()
plt.tight_layout()

# 7. Show plot
plt.show()
```

```python
# 1. Import libraries
import numpy as np
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt

# 2. Create sample data
categories = ['A', 'B', 'C', 'D', 'E']
values = np.random.randint(10, 100, size=5)

# 3. Create a DataFrame
df = pd.DataFrame({
    'Category': categories,
    'Value': values
})

# 4. Create a bar plot using Seaborn
plt.figure(figsize=(8, 5))
sns.barplot(data=df, x='Category', y='Value', palette='viridis')

# 5. Customize the plot
plt.title('Simple Bar Plot')
plt.xlabel('Category')
plt.ylabel('Value')
plt.grid(axis='y', linestyle='--', alpha=0.7)
plt.tight_layout()

# 6. Show the plot
plt.show()
```
