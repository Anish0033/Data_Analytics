```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt

# Create some data using NumPy
x = np.arange(1, 11)  # Numbers from 1 to 10
y = x ** 2            # Square of each number

# Create a Pandas DataFrame
df = pd.DataFrame({
    'X': x,
    'Y': y
})

# Plot using Matplotlib
plt.figure(figsize=(8, 5))
plt.plot(df['X'], df['Y'], marker='o', linestyle='-', color='blue', label='Y = X²')

# Labels and title
plt.xlabel('X values')
plt.ylabel('Y values')
plt.title('Simple Line Plot: Y = X Squared')
plt.legend()
plt.grid(True)
plt.show()
```

```python
subjects=['science','hindi','social science','maths','english','computer science']
marks_1=[65,67,61,47,79,80]
marks_2=[78,67,61,47,79,80]
subjects_len=np.arange(len(subjects))
width=0.4

plt.Figure(figsize=(8,8))
colors=('red','blue','green','yellow','purple','orange')
plt.bar(subjects_len,marks_1, width=width,color=colors,label='student1')
plt.bar(subjects_len+width,marks_2,width=width,color=colors,alpha=0.5,label='student2')

plt.yticks(np.arange(0,100,10))
plt.xticks(subjects_len + width / 2, subjects)

plt.xlabel('subjects')
plt.ylabel('marks')
plt.title('marks of students')
plt.legend()
plt.show()
```
